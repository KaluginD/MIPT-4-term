```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
lambdas = [404.7, 491.6, 546.1, 578]
ns = [1.7178, 1.7315, 1.7272, 1.7209]
```

```python
pol1 = np.polyfit(lambdas, ns, deg=1)
pol2 = np.polyfit(lambdas, ns, deg=2)
x = np.linspace(400, 580, 100)
y1 = pol1[0] * x + pol1[1]
y2 = pol2[0] * x**2 + pol2[1] * x + pol2[2]
```

```python
plt.figure(figsize=(15,10))
plt.plot(lambdas, ns)
plt.plot(lambdas, ns, 'ro')
plt.plot(x, y1)
plt.plot(x, y2)
plt.title('n($\lambda$)')
plt.grid()
plt.show()
```
