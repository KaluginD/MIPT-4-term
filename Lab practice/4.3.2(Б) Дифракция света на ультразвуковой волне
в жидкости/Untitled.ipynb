```python
Y = np.array([31, 64, 81, 108, 149, 184, 221, 251, 272]) * 4
X = np.array(list(range(4, -5, -1)))
pol1 = np.polyfit(X, Y, deg=1)
x = np.linspace(-4.1, 4.1, 100)
y = pol1[0] * x + pol1[1]
print(pol1)
plt.figure(figsize=(10,10))
plt.title('$Y(n)$ при $\\nu = 1,2 \\cdot 10^6$ Гц')
plt.plot(X, Y, 'ro')
plt.plot(x, y)
plt.plot(x, y, label ='k = %.3f, b = %.3f' %(pol1[0], pol1[1]) )
plt.legend()
```
