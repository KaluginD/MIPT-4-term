```python
import numpy as np
import pandas as pd
from scipy import stats as sts
import matplotlib.pyplot as plt
import math
import seaborn as sbn
from sklearn import metrics as mts
import csv
%pylab inline
```

```python
sample = np.genfromtxt('6.csv', delimiter=',')
sample = sample[3:]
lambd = 45
t_0 = 100
t = 50000
```

$\mathsf{E}( N_t | N_s) = \mathsf{E}(N_s + N_t - N_s | N_s) = N_s + \mathsf{E}( N_t - N_s) = N_s + (t-s) / \lambda$

```python
s = np.linspace(0, t, 10000)
N_s = np.array(list(map(lambda s_once: len(sample[np.cumsum(sample) <= s_once]), s)))
plt.figure(figsize=(20, 10))
plt.plot(s, N_s + (t - s)/lambd, label='lambda ={0}'.format(lambd))
plt.legend()
plt.grid(ls=':')
plt.show()
```
